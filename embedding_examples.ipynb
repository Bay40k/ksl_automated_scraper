### Get all results, regardless of sent status, with `only_unsent=False`
Set `only_unsent` to `True` to only return a list of unsent results

```python
import ksl_scrape

search_results_html = await ksl_scrape.get_search_results("ps5")
parsed_results = await ksl_scrape.parse_search_results(search_results_html, only_unsent=False)

for listing in parsed_results:
    print(listing.title)
```
